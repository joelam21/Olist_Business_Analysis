```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# read in data files
df = pd.read_csv('order_items_products.csv')
```

```python
filter = (  df
            .groupby(['product_id'])
            .agg(sales=('price', 'sum'))
            .sort_values(by='sales', ascending=False)
            .head(1000)
            .index
            .tolist()
            )
```

```python
top_products = (  df
            .groupby(['product_id'])
            .agg(sales=('price', 'sum'))
            .sort_values(by='sales', ascending=False)
            .head(1000)
            )

top_products
```

```python
top_products.to_csv('top_products.csv')
```

```python
# Filter the original DataFrame using the list of top product IDs
filtered_df = df[df['product_id'].isin(filter)]

# Display the filtered DataFrame
filtered_df
```

```python
# Generate sample dataset since whole dataset is too big
#sampled_df = df.sample(n=30000, random_state=80)

basket_df = filtered_df.pivot_table(index='order_id',
               columns='product_id', 
               values='customer_id', 
               aggfunc='count', 
               margins=False
               )

basket = basket_df.fillna(0)

basket.head()
```

```python
basket = basket.astype(bool)
```

```python
basket.info()
```

```python
basket.head()
```

```python
# Convert values to 1 and 0
# basket = basket.applymap(lambda x: 1 if x > 0 else 0)

# Apply Apriori algorithm
frequent_itemsets = apriori(basket, min_support=0.0001, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

analysis_export = rules.sort_values(['support', 'confidence', 'lift'], axis = 0, ascending = False).head(10)

analysis_export
```

```python
analysis_export.to_csv('apriori_algo.csv', index=False)
```
